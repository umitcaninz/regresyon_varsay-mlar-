<h1 style="background-color:#3b5998;font-family:newtimeroman;font-size:250%;text-align:center;color:#FFFFFF">POLİNOM REGRESYON 
</h1>

Polinom regresyonu, bağımlı değişken ile bir veya birden fazla bağımsız değişken arasındaki ilişkiyi modellemek için kullanılan bir regresyon analiz yöntemidir. Lineer regresyonun genişletilmiş bir versiyonudur ve doğrusal olmayan ilişkileri ifade etmek için polinom fonksiyonlarını kullanır. Temelde, regresyon denklemine bağımsız değişkenlerin kuvvetleri ve kareleri eklenir.

Y = $\beta_0 + \beta_1X + \beta_2X^2 + \beta_3X^3 + \ldots + \varepsilon$

$\beta_0$ :  Kesim noktası (intercept)

$\beta_1X$ :  Birinci dereceden katsayı (regresyon katsayısı) 

$\beta_2X^2 + \beta_3X^3 + \ldots$ : İkinci, üçüncü ve daha yüksek dereceli katsayılar

$X$ : Bağımsız değişken

$X^2,X^3$ : Bağımsız değişkenin kareleri, küpleri ve daha yüksek dereceleri

Polinom regresyonu, doğrusal regresyonun yeteneklerini genişleterek daha karmaşık ilişkileri ifade edebilir. Bu sayede, veri setindeki eğilimleri daha esnek bir şekilde modelleyebilir. Ancak, polinom regresyonunda polinom derecesinin seçimi önemlidir. Çünkü çok yüksek dereceli polinomlar, modelin aşırı uydurulmasına (overfitting) neden olabilir.

<h1 style="background-color:#FFA500;font-family:newtimeroman;font-size:200%;text-align:left;color:#FFFFFF">Neden Tercih Edilir ?
</h1>


Polinom regresyon, genellikle veri seti üzerinde doğrusal bir ilişkiyi ifade edemeyen, karmaşık, eğri çizgilere sahip olan durumlarda tercih edilir.

1 ) Doğrusal Olmayan İlişkiler

2 ) Yüksek Dereceli İlişkiler

3 ) Aşırı Uyum (Overfitting)

4 ) Dalgalı Veri Dağılımı

5 ) Model Esnekliği İhtiyacı

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:125%;text-align:left;color:#FFFFFF">1)Doğrusal Olmayan İlişkiler:
</h3>


Eğer veri setinizde bağımlı değişken ile bağımsız değişken arasında doğrusal bir ilişki değil de daha karmaşık bir eğri ilişkisi varsa, polinom regresyonu tercih edilebilir. Örneğin, veri setinizin dağılımı parabolik veya kubik bir yapıya sahipse.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:125%;text-align:left;color:#FFFFFF">2)Yüksek Dereceli İlişkiler:
</h3>

Polinom regresyon, doğrusal regresyona kıyasla daha yüksek dereceli ilişkileri ifade edebilir. Eğer veri setinizdeki ilişki daha yüksek dereceli bir polinom tarafından açıklanabiliyorsa, polinom regresyonu kullanmak anlamlı olabilir.


###### Yüksek Dereceli İlişkiler Nasıl Tespit Edilir ?

###### 2.1)Veriyi Görselleştirme:
Veri setinizi bir grafik üzerinde görselleştirin. Eğer doğrusal bir regresyon çizgisi veri noktalarını yeterince iyi açıklayamıyorsa ve bir eğri veya eğrisel bir desen olduğunu düşünüyorsanız, yüksek dereceli bir polinomun daha uygun olabileceğini düşünebilirsiniz.

###### 2.2)Polinom Derecesini Değiştirme: 
Deneme yanılma yöntemiyle, farklı polinom derecelerini modelleyerek performanslarını karşılaştırabilirsiniz. Örneğin, başlangıçta birinci dereceden başlayarak polinom derecesini artırabilir ve her seferinde modelin performansını değerlendirebilirsiniz.
###### 2.3)Regresyon Analizi:
Regresyon analizi sonuçlarını inceleyerek, modelin belirli bir derecesinin veriyi daha iyi açıkladığına dair istatistiksel anlamlılığı değerlendirebilirsiniz. Bu analizde, belirli bir polinom derecesi için regresyon katsayılarına ve p-değerlerine dikkat edebilirsiniz.
###### 2.4)Model Uyumunu Değerlendirme: 
Belirli bir polinom derecesi için modelin eğitim seti üzerindeki hatasını ve test seti üzerindeki hatasını değerlendirin. Model, eğitim setinde çok iyi performans gösterirken test setinde düşük performans gösteriyorsa, aşırı uyum olabilir.
###### 2.5)AIC veya BIC Kriterleri: 
AIC (Akaike Bilgi Kriteri) veya BIC (Bayesian Bilgi Kriteri) gibi bilgi kriterlerini kullanarak farklı polinom dereceleri arasında bir seçim yapabilirsiniz. Daha düşük AIC veya BIC değerleri, modelin daha iyi uyduğunu gösterebilir.

Bu adımları uygulayarak, veri setinizin doğasını daha iyi anlayabilir ve en uygun polinom derecesini belirleyebilirsiniz. Ancak, dikkat etmeniz gereken önemli bir nokta, aşırı uyuşma riskidir. Çok yüksek dereceli polinomlar genellikle eğitim setine aşırı uyarlar ve genelleme yetenekleri düşer. Bu nedenle, model karmaşıklığını dikkatlice değerlendirmek önemlidir.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:125%;text-align:left;color:#FFFFFF">3)Aşırı Uyum (Overfitting):
</h3>

Doğrusal regresyon modeli veri setini kötü bir şekilde uyuyorsa (underfitting) veya tersine çok fazla ayrıntıya odaklanıp veri setine aşırı uyum sağlıyorsa (overfitting), polinom regresyonu daha iyi bir sonuç verebilir. Ancak, aşırı uyum riskine karşı dikkatli olunmalıdır.


#### Aşırı Uyum Nasıl Tespit Edilir ?
###### 3.1)Eğitim ve Test Hata Analizi: 

Modelinizi eğittikten sonra, hem eğitim veri seti üzerinde hem de ayrılmış bir test veri seti üzerinde performansını değerlendirin. Eğitim hatası düşük olabilir, ancak test hatası yüksekse, aşırı uyum olabilir.


###### 3.2)Çapraz Doğrulama (Cross-Validation): 

K-fold çapraz doğrulama kullanarak modelinizi farklı alt kümelerde eğitip değerlendirin. Eğer eğitim hatası ile test hatası arasında büyük bir fark varsa, aşırı uyum olabilir.


###### 3.3)Öğrenme Eğrisi (Learning Curve) Analizi: 

Eğitim ve test setlerinin boyutlarına karşı hata oranlarını inceleyerek modelinizi değerlendirin. Eğer model eğitim verisiyle hızla öğreniyorsa ancak test setinde performans düşükse, aşırı uyum olabilir.

###### 3.4)Model Karmaşıklığı Kontrolü: 

Polinom derecesi gibi model parametrelerini kontrol edin. Model karmaşıklığını azaltmak, örneğin polinom derecesini düşürmek, aşırı uyuşmayı önleyebilir.

###### 3.5)Regülerizasyon Kullanımı: 

Regülerizasyon terimleri eklemek, özellikle L1 veya L2 regülerizasyonu, modelin aşırı uyum eğilimini azaltabilir.

###### 3.6)Veri Seti Büyüklüğünü Artırma: 

Daha fazla veri kullanmak, modelin genelleme yeteneğini artırabilir ve aşırı uymayı azaltabilir.

###### 3.7)Farklı Model Denemeleri: 

Farklı karmaşıklık seviyelerindeki modelleri deneyin. Örneğin, polinom derecesini değiştirerek modeli basitleştirebilir veya karmaşıklığı artırabilirsiniz.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:200%;text-align:center;color:#FFFFFF"> Polinom Regresyonunda Polinom Derecesi Seçimi
</h3>

Polinom regresyon, modelin esnekliğini artıran bir yöntemdir. Eğer veri setinizdeki ilişki karmaşık ve değişken ise, polinom regresyonu daha esnek bir model sunabilir.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">1)R-Kare Değeri
</h3>

R-kare değeri, bir regresyon modelinin bağımlı değişkenin varyansının ne kadarını açıkladığını gösterir. Ancak, R-kare değerinin tek başına bir modelin başarısını belirlemedeki etkinliği sınırlıdır. Özellikle yüksek dereceli polinomlarda, R-kare değeri yüksek olabilir, ancak model gerçek verilere aşırı uymuş (aşırı uyuma) olabilir.

Yüksek bir R-kare değeri genellikle iyi bir model performansını gösterir, ancak bu durumun aşırı uyuma neden olup olmadığını değerlendirmek için diğer faktörlere de bakmak önemlidir. Özellikle polinom regresyonunda, polinom derecesi arttıkça modelin eğitim verilerine daha iyi uyması, ancak genelleme yeteneğinin düşmesi muhtemeldir.

###### R-kare Değeri 0.7-0.8:
İyi bir model performansını gösterir ve genellikle kabul edilebilirdir.

###### R-kare Değeri 0.8-0.9:
Çok yüksek bir R-kare değeri, modelin veriyi aşırı uydurmuş olabileceği anlamına gelebilir. Dikkatli bir şekilde incelenmeli ve diğer değerlendirme kriterleri ile birlikte değerlendirilmelidir.

###### R-kare Değeri 0.9 ve Üzeri:
Yüksek R-kare değerleri, genellikle aşırı uyuma işaret eder. Bu durumda, modelin genelleme yeteneğini değerlendirmek için test verileri veya çapraz doğrulama kullanılmalıdır.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:250%;text-align:left;color:#FFFFFF">2)Çapraz Doğrulama (Cross Validation): 
</h3>

###### Nedir ?

Çapraz doğrulama (cross-validation), bir modelin genelleme yeteneğini değerlendirmek için kullanılan önemli bir tekniktir. Polinom regresyonu gibi esnek modellerde, çapraz doğrulama, farklı polinom dereceleri için model performansını değerlendirmek için özellikle önemlidir.


<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">2.1)Çapraz Doğrulama (Cross Validation): 
</h3>

##### Veri Seti Bölme ve K-Fold Çapraz Doğrulama:
Eğitim ve test seti oranlarını belirlerken genellikle problem bağlamına, veri setinizin boyutuna ve genel amacınıza göre bir denge bulmanız gerekmektedir.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">2.2) Klasik Oranlar:
</h3>

###### Nasıl Belirlenir:
Klasik olarak kullanılan %80 eğitim ve %20 test oranları genelde orta büyüklükteki veri setleri için başlangıç noktasıdır.
Ancak, büyük veri setlerinde %90 eğitim ve %10 test gibi daha küçük bir test seti oranı kullanılabilir.
Bu oranlar, veri setinizin doğasına ve problem bağlamına bağlı olarak değişebilir.
###### Örnek:
Örneğin, 100,000 örnek içeren bir veri setiniz varsa, başlangıç ​​olarak %80 eğitim ve %20 test oranı uygun olabilir.
Ancak, 1,000,000 örnek içeren bir büyük veri setiniz varsa, %90 eğitim ve %10 test oranı daha etkili olabilir.
###### Avantajlar:
Yaygın olarak kullanılır ve basittir.
Başlangıç noktası olarak kullanılabilir.
###### Dezavantajlar:
Veri setinizin özelliklerine bağlı olarak bu oranlar uygun olmayabilir.



<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF"> 2.3)K-Stratified Cross-Validation:
</h3>

##### K-Stratified Cross-Validation:
###### Nedir ?
Veri seti k katlamalı çapraz doğrulama ile bölünür ve her katlama içinde oranlar dengelenir.
###### Nasıl Belirlenir:
K-Stratified Cross-Validation, sınıf dengesini korumak için kullanılır. Özellikle küçük veri setlerinde sınıf dengesini korumanın önemi büyüktür.
K değeri, kullanıcının belirlediği katlama sayısını temsil eder. Genellikle 5 veya 10 kullanılır.
###### Örnek:
Diyelim ki, 5 katlamalı K-Stratified Cross-Validation kullanacaksınız. Eğitim için %80, test için %20 oranını koruyarak veriyi 5 kere bölüp değerlendirirsiniz.
###### Katkıları: 
K-Stratified Cross-Validation, veri setini k katlamalı çapraz doğrulama ile bölerek her katlamada sınıf dağılımını korur. Bu, özellikle sınıflar arasında dengesizlik olduğunda modelin daha dengeli bir şekilde değerlendirilmesine olanak tanır.
###### Nasıl Çalışır: 
Veri seti, k adet alt küme veya "katlama"ya bölünür. Her katlama sırasında, tüm sınıfların temsil edildiği ve sınıf oranlarının korunduğu bir alt küme oluşturulur. Model, kere kere eğitilir ve test edilir, her seferinde farklı bir katlama kullanılır.
###### Hangi Durumda Kullanılır: 
Özellikle sınıflar arasında dengesizlik olduğunda veya sınıfların eşit temsil edilmesi gerektiğinde tercih edilir.
###### Avantajlar:
Sınıf dengesini koruyarak çapraz doğrulama yapar.
Özellikle sınıf dengesi kritikse tercih edilir.
###### Dezavantajlar:
Küçük veri setlerinde kullanılabilirlik sorunu olabilir.



<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">2.4) Problem Bağlamına Göre Ayarlama:
</h3>

###### Nedir ?
Probleminize bağlı olarak, daha büyük bir test seti kullanmanız gerekebilir. Özellikle nadir sınıfların olduğu durumlarda daha büyük bir test seti, modelin nadir durumları daha etkili bir şekilde öğrenmesine olanak tanıyabilir.
###### Katkıları: 
Bu yöntem, özellikle nadir sınıfların olduğu durumlarda daha büyük bir test seti kullanarak, modelin nadir durumları daha etkili bir şekilde öğrenmesine olanak tanır.
###### Nasıl Çalışır: 
Test seti genellikle daha büyük bir oranda ayrılır, böylece model nadir durumları daha etkili bir şekilde öğrenebilir. Bu durum, nadir olayları tespit etmek veya modelin nadir durumlar üzerindeki performansını artırmak istendiğinde kullanılabilir.
###### Hangi Durumda Kullanılır: 
Nadir durumların olduğu, sınıflar arasında dengesizlik olduğu veya belirli durumların öncelikli olduğu durumlarda tercih edilebilir.
###### Nasıl Belirlenir:
Probleminize bağlı olarak, daha büyük bir test seti kullanmanız gerekebilir. Özellikle nadir durumları etkili bir şekilde öğrenmek istiyorsanız test setini artırabilirsiniz.
###### Örnek: 
Nadir bir hastalığın teşhisini yapmak istiyorsanız, test setinizi büyüterek nadir durumu daha etkili bir şekilde öğrenebilirsiniz.
###### Avantajlar:
Nadir durumları daha etkili bir şekilde öğrenmek için uygundur.
###### Dezavantajlar:
Genel bir kılavuz olmadığı için deneme yanılma gerektirebilir.


<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">2.5) Grid Search veya Random Search ile Deneme:
</h3>

###### Nedir ? 
Makine öğrenimi modelleri genellikle hiperparametre ayarlaması gerektirir. Bu nedenle, farklı oranları denemek için bir grid search veya random search yapabilirsiniz. Bu yöntemle farklı oranları deneyerek model performansını ölçebilir ve en iyi oranı seçebilirsiniz.
###### Katkıları: 
Bu yöntem, hiperparametre ayarlaması yaparken farklı eğitim ve test seti oranlarını denemenizi sağlar. Modelin performansını optimize etmek için oranları sistemli bir şekilde keşfetmenize yardımcı olabilir.
###### Nasıl Çalışır: 
Bir parametre aralığı belirlenir (örneğin, eğitim seti için %70 ila %90 arasında değişen oranlar). Ardından, bu oranları içeren farklı parametre kombinasyonları denemek için grid search veya random search yapılır.
###### Hangi Durumda Kullanılır: 
Modelin performansını en üst düzeye çıkarmak ve hiperparametrelerin doğru ayarlarını bulmak istendiğinde kullanılır.
###### Nasıl Belirlenir:
Hiperparametre ayarlaması yaparken farklı oranları denemek için grid search veya random search kullanılabilir.
###### Avantajlar:
Esnek bir yaklaşım sunar, hiperparametre optimizasyonu için kullanışlıdır.


<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">2.6) Dengeli Sınıf Dağılımı:
</h3>

###### Nedir ?
Sınıflar arasında dengesizlik varsa, sınıf oranlarına dikkat edilmelidir. Her sınıftan eşit sayıda örnek içeren bir test seti, dengesiz sınıfları içeren modellerin performansını daha iyi değerlendirebilir.
###### Katkıları: 
Sınıflar arasındaki dengesizlik varsa, her sınıftan eşit sayıda örnek içeren bir test seti oluşturarak modelin dengesiz sınıfları daha iyi değerlendirmesine olanak tanır.
###### Nasıl Çalışır: 
Test seti oluşturulurken, her sınıftan eşit sayıda örnek içeren bir test seti seçilir. Bu, modelin her sınıfın performansını dengeli bir şekilde değerlendirmesine yardımcı olabilir.
###### Hangi Durumda Kullanılır: 
Sınıflar arasında belirgin bir dengesizlik olduğu ve modelin her sınıfın performansını dengeli bir şekilde değerlendirmesi gerektiği durumlarda tercih edilir.
###### Avantajlar:
Sınıf dengesizliğini ele alarak adil bir performans değerlendirmesi sağlar.
Dengesiz veri setlerinde daha güvenilirdir.
###### Dezavantajlar:
Her zaman mümkün olmayabilir, özellikle sınıf sayısı çok fazlaysa.
###### Nasıl Belirlenir:
Sınıf dengesizliği varsa, her sınıftan eşit sayıda örnek içeren bir test seti oluşturulabilir.
###### Örnek:
Diyelim ki, iki sınıfınız var ve birinci sınıftan çok daha fazla örnek var. Test seti oluştururken her iki sınıftan eşit sayıda örnek içermeye özen gösterirsiniz.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">3)Hata Metrikleri:
</h3>

MSE (Ortalama Kare Hatası) veya MAE (Ortalama Mutlak Hata) gibi hata metrikleri kullanarak farklı polinom dereceleri için model performansını değerlendirin. Model karmaşıklığı ile hata arasındaki dengeyi bulmaya çalışın.

###### Ortalama Kare Hatası (MSE):
MSE = $\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
###### Kullanım Amaçları:
Büyük hatalara daha fazla vurgu yapmak istendiğinde.
Aykırı değerlere (outliers) duyarlı olmayan bir metrik istendiğinde.
###### Avantajlar:
Aykırı değerlere duyarlıdır, bu nedenle genellikle gerçek dünyadaki uygulamalarda tercih edilir.
###### Dezavantajlar:
Aykırı değerlere hassas olabilir, bu nedenle bu tür durumlarla başa çıkmak zor olabilir.
###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 


###### Ortalama Mutlak Hata (MAE):
MAE = $\frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$

###### Kullanım Amaçları:
Büyük ve küçük hatalara eşit vurgu yapmak istendiğinde.
Aykırı değerlere dirençli bir metrik istendiğinde.
###### Avantajlar:
Aykırı değerlere dirençlidir, bu nedenle genelde güvenilir bir performans ölçütüdür.
###### Dezavantajlar:
Büyük hatalara daha az vurgu yapar, bu nedenle gerçek dünyadaki durumları tam olarak yansıtmayabilir.



#### Karar Verme:

Eğer modelinizin performansının genel bir ölçümünü istiyorsanız, R-kare tercih edilebilir.
MSE ve MAE, özellikle belirli hata türlerine öncelik vermek istiyorsanız veya modelinizin hassasiyetini belirli bir aralığa odaklamak istiyorsanız kullanılabilir.
Aykırı değerlere karşı daha dirençli bir metrik isteniyorsa MAE tercih edilebilir.

<h3 style="background-color:#FF00F7;font-family:newtimeroman;font-size:150%;text-align:left;color:#FFFFFF">4)AIC ve BIC
</h3>

Akaike Bilgi Kriteri (AIC) ve Bayesian Bilgi Kriteri (BIC), model seçimi için kullanılan istatistiksel bilgi kriterleridir. Bu kriterler, model karmaşıklığı ve veriye uyumun bir dengelemesini sağlamak amacıyla kullanılır. AIC ve BIC arasındaki temel fark, BIC'nin daha fazla kısıtlamaya ve daha büyük veri setlerine odaklanan bir düzeltme içermesidir.

###### Akaike Bilgi Kriteri (AIC):
AIC, modelin performansını değerlendirmek için kullanılan bir bilgi kriteridir. AIC, modelin uyumunu ölçerken aynı zamanda modelin karmaşıklığını da penalize eder. AIC'nin düşük olması, daha iyi bir modelin işaretidir.

###### AIC=2k−2ln(L)

k : Modelin serbest parametre sayısı (modelin karmaşıklığı).

ln(L) : Maksimum olabilirlik fonksiyonunun doğal logaritması.



###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 


###### Bayesian Bilgi Kriteri (BIC):
BIC de AIC gibi model seçiminde kullanılan bir kriterdir. Ancak, BIC, daha büyük veri setleri ve daha fazla parametre içeren modeller için daha fazla düzeltme içerir. BIC'nin temel amacı, karmaşık modelleri daha güçlü bir şekilde penalize etmektir.

BIC=kln(n)−2ln(L)

n: Veri setinin örnek sayısı.

k: Modelin serbest parametre sayısı (modelin karmaşıklığı).

ln(L) : Maksimum olabilirlik fonksiyonunun doğal logaritması.
###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 


###### Model Seçimi:

AIC ve BIC değerleri karşılaştırılarak, daha küçük olan değerler tercih edilir.
Daha küçük AIC veya BIC değerine sahip bir model, veriyi daha iyi açıklayabilir ve daha düşük karmaşıklığa sahip olabilir.

###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 


###### Dikkat Edilmesi Gereken Noktalar:

AIC ve BIC, sadece modeller arasında karşılaştırma yapmak için kullanılır. Tek başına bir modelin kalitesini ölçmezler.
İki kriter de düşük değere sahip bir modelin tercih edilmesi gerektiğini gösterir, ancak hangi kriterin kullanılacağı probleme bağlıdır.

### Polinom regresyon analizi sırasında ANOVA dışında değerlendirilebilecek birkaç önemli istatistiksel ölçüt ve yorumlama adımı bulunmaktadır. 

##### Model Katsayıları (Coefficients):

Polinom regresyon modelindeki katsayılar, bağımsız değişkenlerin modeldeki etkisini gösterir. Her bir katsayının anlamlılığını değerlendirmek için genellikle t-testi veya p-değerleri kullanılır. Anlamsız katsayılar, modelde o belirli terimi kullanmanın anlamsız olduğunu gösterebilir.

######   Anlamsız Katsayılar:

Yorum: Eğer polinom regresyon modelindeki bir katsayı anlamsız (p-değeri yüksek) ise, ilgili bağımsız değişkenin modeldeki etkisi istatistiksel olarak anlamlı değildir.

Etki: Anlamsız katsayılar, modelin açıklama gücünü düşürebilir ve ilgili bağımsız değişkenin modeldeki rolünü sorgulatır.

Ne Yapılır ? : Anlamsız katsayılara sahip değişkenler modelden çıkarılabilir veya modelin tekrar değerlendirilmesi için alternatif yöntemlere başvurulabilir.



###### Anlamlı Katsayılar:

Yorum: Eğer polinom regresyon modelindeki bir katsayı anlamlı (p-değeri düşük) ise, ilgili bağımsız değişkenin modeldeki etkisi istatistiksel olarak anlamlıdır.

Etki: Anlamlı katsayılar, modelin açıklama gücünü artırabilir ve ilgili bağımsız değişkenin modeldeki önemini vurgular.

Ne Yapılır ? : Anlamlı katsayılara sahip değişkenler, modelin güvenilirliğini artırmak için korunabilir ve yorumlanabilir.
###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 


###### Model İstatistikleri:

###### F İstatistiği ve p-değerleri:

Yorum: F istatistiği, modelin genel anlamda anlamlılığını değerlendirir. Anlamlı bir F istatistiği, en az bir bağımsız değişkenin modelde anlamlı bir etkiye sahip olduğunu gösterir. P-değerleri, F istatistiğinin anlamlılığını belirler; düşük p-değerleri genellikle anlamlı bir model önerir.

Etki: Anlamlı F istatistiği ve düşük p-değerleri, modelin genel performansının anlamlı olduğunu ve en az bir bağımsız değişkenin etkili olduğunu gösterir.

Ne Yapılır ? : Modelin genel anlamlılığını değerlendirmek için F istatistiği ve p-değerleri kullanılabilir. Anlamlı bir model, ileri analizlere geçmek için bir temel sağlar. P-değerleri, hangi bağımsız değişkenlerin anlamlı olduğunu belirlemek için kullanılabilir.
###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 

###### R-kare:

Yorum: R-kare, bağımlı değişkenin varyansının ne kadarının açıklandığını gösterir. Yüksek R-kare değerleri, modelin veriyi iyi açıkladığını gösterir. Düşük R-kare değerleri, modelin zayıf bir uyum sağladığını gösterebilir.

Etki: Yüksek R-kare değerleri, modelin genel performansının iyi olduğunu ve bağımsız değişkenlerin bağımlı değişkenin varyansını açıklamada etkili olduğunu gösterir. Düşük R-kare değerleri, modelin açıklama gücünün sınırlı olduğunu gösterir.

Ne Yapılır ? : Modelin açıklama gücünü değerlendirmek için R-kare kullanılabilir. Ancak, R-kare tek başına yeterli bir değerlendirme ölçütü değildir; modelin diğer yönleri de göz önüne alınmalıdır.
###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 

###### Eksik veya Düşük Anlamlılık:

Yorum: Eğer F istatistiği anlamsızsa veya p-değerleri yeterince düşük değilse, modelin genel anlamda anlamlı olmadığı söylenebilir. Bu durumda, modelin önerdiği bağımsız değişkenlerin bağımlı değişken üzerinde anlamlı bir etkisi olmayabilir.

Etki: Anlamsız bir model, ileri analizlere geçmeden önce modelin gözden geçirilmesini gerektirir.

Ne Yapılır ? : Modelin genel anlamlılığını belirlemek için F istatistiği ve p-değerleri değerlendirilmelidir. Anlamsız bir model, değişken seçimi, model değişiklikleri veya farklı regresyon yöntemleri gibi aksiyonları gerektirebilir.

###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 

##### Heteroskedastisite Kontrolü:

Polinom regresyon modellerinde hata terimlerinin (artıkların) heteroskedastisite gösterip göstermediğini değerlendirmek önemlidir. Heteroskedastisite, artıkların modelin farklı bölgelerinde farklı varyanslara sahip olması durumunu ifade eder. Bu durum, modelin güvenilirliğini etkileyebilir.


Yorum: Heteroskedastisite, artıkların farklı bölgelerde farklı varyanslara sahip olması durumunu ifade eder. Bu durum, modelin güvenilirliğini etkileyebilir.

###### Varyans Homojen Değilse:

Yorum: Artık grafikleri veya heteroskedastisite testleri, artıkların varyansının homojen olmadığını gösteriyorsa, modelin güvenilirliği konusunda sorunlar ortaya çıkabilir. Heteroskedastisite, özellikle standart hata tahminlerini yanıltabilir.

Etki: Heteroskedastisite, modelin güvenilirliğini azaltabilir ve parametre tahminlerinin doğruluğunu etkileyebilir. 
Heteroskedastisite durumunda, modelin güvenilirliğini değerlendirmek zorlaşır.

Ne Yapılır ? : Eğer heteroskedastisite belirlenirse, regresyon modelinin gözden geçirilmesi ve düzeltilmesi gerekir. Bu düzeltmeler arasında dönüşümler, ağırlıklandırılmış en küçük kareler (WLS), veya heteroskedastisite-robust standart hataların kullanılması yer alabilir.

###### Varyans Homojen İse:

Yorum: Eğer artıkların varyansı homojen ise, modelin güvenilirliği artar. Bu durum, regresyon katsayılarının tahminlerinin daha güvenilir olduğunu gösterir.

Etki: Homojen varyans, modelin daha tutarlı ve güvenilir tahminler sağlamasına yardımcı olur.

Ne Yapılır ? : Homojen varyans durumunda, modelin güvenilirliği konusunda belirgin bir endişe yoksa, modelin kullanımına devam edilebilir. Ancak, her durumda homojen varyansın diğer model değerlendirme kriterleri ile birlikte değerlendirilmesi önemlidir.

###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 

##### Artıkların Normal Dağılımı:

Modelin hatası, genellikle normal bir dağılıma sahip olmalıdır. Normalite testleri veya artıkların histogramları incelenerek bu özellik kontrol edilebilir. Normalite ihlalleri, modelin güvenilirliğini etkileyebilir.

###### Normal Dağılıyor İse:

Yorum: Eğer artıkların histogramı veya normalite testleri normal bir dağılımı gösteriyorsa, bu durum modelin güvenilirliği açısından olumlu bir işarettir.

Etki: Normal dağılıma uygunluk, parametre tahminlerinin güvenilir olmasını sağlar ve istatistiksel testlerin doğruluğunu artırır.

Ne Yapılır ?: Eğer artıklar normal dağılıyorsa, modelin normalite açısından bir sorunu olmadığından emin olabilirsiniz. Bu durumda modelin kullanımına devam edilebilir.

###### Normal Dağılım Göstermiyorsa:

Yorum: Eğer artıkların histogramı veya normalite testleri normal bir dağılımı göstermiyorsa, bu durum modelin güvenilirliği açısından endişe yaratabilir.

Etki: Normal dağılıma uygunluk eksikliği, parametre tahminlerinin ve hipotez testlerinin güvenilirliğini azaltabilir.

Ne Yapılır ? : Normal dağılım ihlali durumunda, dönüşümler veya farklı regresyon yöntemleri gibi düzeltme yöntemleri düşünülmelidir.Alternatif olarak, büyük örneklem boyutlarıyla normal dağılıma uygunluğun sağlanıp sağlanmadığı tekrar değerlendirilebilir.

###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 

##### Durbin-Watson İstatistiği:

Durbin-Watson istatistiği, artıklar arasındaki otokorelasyonu değerlendirir. Bu, artıkların zaman içinde birbirleriyle ilişkili olup olmadığını kontrol etmeye yardımcı olabilir. Değer 2'ye yakınsa, otokorelasyon olasılığı düşüktür.

###### Otokorelasyon Var İse:

Yorum: Durbin-Watson istatistiği değeri 2'den küçükse, artıklar arasında pozitif otokorelasyon olasılığı yüksektir. Bu durum, modelin güvenilirliğini etkileyebilir.

Etki: Otokorelasyon, artıkların bağımsız olmadığını ve modelin tahmin gücünü azaltabileceğini gösterir.

Ne Yapılır ?: Eğer otokorelasyon varsa, modelin güvenilirliğini artırmak için alternatif regresyon teknikleri veya dönüşümler düşünülmelidir.

###### Otokorelasyon Yok İse:

Yorum: Durbin-Watson istatistiği değeri 2'ye yakınsa veya büyükse, artıklar arasında otokorelasyon olasılığı düşüktür. Bu durum model açısından olumlu bir durumdur.

Etki: Otokorelasyonun olmaması, artıkların bağımsız olduğunu ve modelin tahmin gücünü koruduğunu gösterir.

Ne Yapılır ? : Bu durumda, otokorelasyon nedeniyle bir müdahalede bulunmaya gerek yoktur. Modelin kullanımına devam edilebilir.



###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 


##### VIF (Variance Inflation Factor):

Variance Inflation Factor (VIF), çoklu doğrusal regresyon modellerindeki bağımsız değişkenler arasındaki korelasyonu ölçen bir istatistiksel gösterge olarak kullanılır. Yüksek VIF değerleri, bağımsız değişkenler arasında güçlü bir ilişkinin varlığını gösterir ve bu durum regresyon analizinde tahmin güvenilirliğini etkileyebilir.

###### VIF Değerleri Yüksek İse (Genellikle 10'dan Büyük):

Yorum: Yüksek VIF değerleri, bağımsız değişkenler arasında güçlü bir ilişkinin veya çoklu doğrusallığın olduğunu gösterir.

Etki: Çoklu doğrusallık, modelin güvenilirliğini azaltabilir ve katsayıların yorumlanabilirliğini zorlaştırabilir.

Ne Yapılır ? : Yüksek VIF değerlerini çözmek için, ilişkilendirilmiş bağımsız değişkenler arasından birini çıkarmak veya değişkenleri dönüştürmek gibi yöntemler düşünülebilir.


###### VIF Değerleri Kabul Edilebilir İse (Genellikle 10'dan Küçük veya 5'ten Küçük):

Yorum: Kabul edilebilir düzeydeki VIF değerleri, bağımsız değişkenler arasında anlamlı bir çoklu doğrusallık belirtisi olmadığını gösterir.

Etki: Modelin güvenilirliği ve katsayıların anlamlılığı açısından olumlu bir durumdur.

Ne Yapılır ?: Bu durumda özel bir müdahale gerekmez ve model kullanılmaya devam edilebilir.

###### _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ 

#### Model Değerlendirme ve Doğrulama:

Modelin test verileri üzerinde nasıl performans gösterdiğini değerlendirmek önemlidir. Çapraz doğrulama ve diğer model değerlendirme teknikleri, modelin genelleme yeteneğini kontrol etmeye yardımcı olabilir. 

###### Genelleme Yeteneği Yoksa:

Yorum: Eğer model, test verileri üzerinde iyi bir performans göstermiyorsa, genelleme yeteneği düşük olabilir. Bu durum, modelin eğitim verilerine aşırı uyum (overfitting) yaptığını veya modelin karmaşıklığının fazla olduğunu gösterebilir.

Etki: Modelin pratikte kullanılabilirliği düşer ve yeni, görünmeyen verilere adapte olma kabiliyeti azalır.

Ne Yapılır ? : Modelin karmaşıklığını azaltmak için düzenleme teknikleri uygulanabilir, daha fazla veri toplanabilir veya başka bir modelleme yaklaşımı düşünülebilir.

###### Genelleme Yeteneği Varsa: 

Yorum: Model, test verileri üzerinde iyi bir performans gösteriyorsa, genelleme yeteneği yüksektir. Bu durum, modelin eğitim verilerinden bağımsız verileri iyi tahmin edebildiğini gösterir.

Etki: Modelin pratikte kullanılabilirliği artar ve yeni verilere adapte olma kabiliyeti güçlenir.

Ne Yapılır ?: Modelin performansını korumak için düzenleme teknikleri ve doğrulama stratejileri devam ettirilebilir.
